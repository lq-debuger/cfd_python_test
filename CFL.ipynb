# Convergence and the CFL Condition

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def LinearCov(nx):
    dx = 2 / (nx -1)
    nt = 20
    c =  1
    sigma = .5
    dt = sigma * dx

    u = np.ones(nx)
    u[int(.5/dx):int(1/dx +1)] = 2
    
    un = np.ones(nx)

    for n in range(nt):
        un = u.copy()
        for i in range(1,nx):
            u[i] = un[i] -c*dt/dx*(un[i]-un[i-1])
    
    plt.plot(np.linspace(0,2,nx),u)
```

```python
LinearCov(41)
```

```python
LinearCov(70)
```
